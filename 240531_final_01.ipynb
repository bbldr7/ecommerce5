```python
# 1) order_number = 1 

# 첫 방문 고객 식별
first_time_customers = df[df['order_number'] == 1]

# 첫 방문 고객 중 유기농 제품을 구매한 수
first_time_organic = first_time_customers[first_time_customers['is_organic'] == True]['user_id'].nunique()

# 첫 방문 고객 수
total_first_time_customers = first_time_customers['user_id'].nunique()

# 전체 유기농 제품 구매 고객 수
total_organic_customers = df[df['is_organic'] == True]['user_id'].nunique()

# 전체 고객 수
total_customers = df['user_id'].nunique()

# 비율 계산
first_time_organic_ratio = first_time_organic / total_first_time_customers if total_first_time_customers else 0
overall_organic_ratio = total_organic_customers / total_customers if total_customers else 0

(first_time_organic_ratio, overall_organic_ratio)

```
