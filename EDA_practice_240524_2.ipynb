```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
orders = pd.read_csv('/content/drive/MyDrive/프로젝트_파이썬/orders.csv')
orders
```

```python
products = pd.read_csv('/content/drive/MyDrive/프로젝트_파이썬/products.csv')
products
```

```python
opp = pd.read_csv('/content/drive/MyDrive/프로젝트_파이썬/order_products__prior.csv')
```

```python
opp_m = pd.merge(opp, products, on='product_id', how='left' )
```

```python
dpt = pd.read_csv('/content/drive/MyDrive/프로젝트_파이썬/departments.csv')
```

```python
opp_m2 = pd.merge(opp_m, dpt, on = 'department_id', how='left')
opp_m2
```

```python
opt = pd.read_csv('/content/drive/MyDrive/프로젝트_파이썬/order_products__train.csv')
```

```python
# prior, train 합친 거

opp1 = opp.copy()
opt1 = opt.copy()
oppt = pd.concat([opp1, opt1])

oppt = oppt.reset_index(drop=True)
```

```python
oppt_m = pd.merge(oppt, products, on='product_id', how='left' )
```

```python
oppt_m2 = pd.merge(oppt_m, dpt, on = 'department_id', how='left')
oppt_m2
```

```python
orders.info()
```

```python
# 1. 전체 주문 수

orders['order_id'].nunique()
```

```python
# 2. 구매자 수

orders['user_id'].nunique()
```

```python
# 3. 상품 별 주문 건수 (prior만)

opp.groupby('product_id')['order_id'].count()
```

```python
# 3-2. 상품 별 주문 건수 (두개 합친 거)

oppt.groupby('product_id')['order_id'].count()
```

```python
# 4-1. 이용자별 주문 수

orders.groupby('user_id')[['order_id']].count()
```

```python
#4-2. 이용자별 평균 주문 수

order_counts = orders.groupby('user_id')[['order_id']].count()

order_counts.mean()
```

```python
# 번외) 주문자별 최소 주문 수

order_count_min = order_counts.min()
order_count_min
```

```python
# 5. 가장 많이 판매되는 상품 10위

# prior만

opp_m.groupby(['product_id','product_name'])['order_id'].count().sort_values(ascending=False).head(10)
```

```python
# 두개 합친 거

oppt_m.groupby(['product_id','product_name'])['order_id'].count().sort_values(ascending=False).head(10)
```

```python
# 6. 카트에 첫 번째로 담기는 상품 10개

['add_to_cart_order == 1']
```

```python
#7. 시간별 주문 건수

orders_per_hour = orders.groupby('order_hour_of_day')['order_id'].count()

orders_per_hour.sort_values(ascending=False).head(10)

# => 오전, 오후 혹은 시간 블록으로 나눠보는 것도 좋을 거 같다
```

```python
# 8. 첫 구매 후 다음 구매까지 걸린 평균 일수

orders['days_since_prior_order'].mean()
```

```python
# 9. 재주문율이 높은 상품

# prior만

opp_m.groupby(['product_id', 'product_name'])['reordered'].count().sort_values(ascending=False).head(10)
```

```python
# 두개 합친 거

oppt_m.groupby(['product_id', 'product_name'])['reordered'].count().sort_values(ascending=False).head(10)
```

```python
# 10. Department 별 재구매율이 높은 상품

opp_m.groupby('department_id')['reordered'].count().sort_values(ascending=False).head(10)
```
